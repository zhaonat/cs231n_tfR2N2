```python
#get an idea of how encoder decoder models should work in tensorflow
import numpy as np
import tensorflow as tf
```

```python
time_steps = 10; #length of sequence
n_input = 20; #dimensionality of each input at each time step
tf.reset_default_graph();
x = tf.placeholder(tf.float32, shape = [None, time_steps, n_input])

#the input that we send must be a set of inputs, one for each time step 

input=tf.unstack(x ,time_steps,1)

#encoder
lstm_size = 10; #10 cells...so up to 10 outputs
encoder = tf.contrib.rnn.BasicLSTMCell(lstm_size)

#to run, you use static run
outputs, _ = tf.contrib.rnn.static_rnn(encoder, input, dtype = tf.float32)

#for each time step, we output y_t, each y_t has some dimension... of size batch?

init = tf.global_variables_initializer();
sess = tf.Session();
sess.run(init);
test_x = np.random.rand(14, 10, 20);
test_out = sess.run(outputs, feed_dict = {x: test_x})
print(test_out)

## decoder
#apparently for machine translation, the decoder is also just an LSTM, take outputs of first lstm
# and feed it through... seems redundant...

decoder = tf.contrib.rnn.BasicLSTMCell(lstm_size)

```
